```python
import spacy
nlp = spacy.load('en_core_web_lg')
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation

skip_words = list(STOP_WORDS) + list(punctuation) + ['\n']
text = open("text.txt", "r", encoding="utf8").read()
doc = nlp(text)

all_words = [word.text for word in doc]

#print(all_words)

# create hash
wordFreq = {}
for w in all_words:
    word = w.lower()
    if word not in skip_words and word.isalpha():
        if word not in wordFreq.keys():
            wordFreq[word] = 1
        else:
            wordFreq[word] += 1

#print(wordFreq)
```

```python
maxFreq = max(wordFreq.values())

# Normalisera listan
for word in wordFreq.keys():
    wordFreq[word] = (wordFreq[word]/maxFreq) # n/max, 12/233
    
#print(wordFreq)


```

```python
sentStrength = {}

for sent in doc.sents:
    for word in sent:
        if word.text.lower() in wordFreq.keys():
            # hash it
            if sent not in sentStrength.keys():
                sentStrength[sent] = 0
            sentStrength[sent] += wordFreq[word.text.lower()]

# print(sentStrength)
 

   
```

```python
sortedSents = sorted(sentStrength.values())[::-1]
topPercentageSents = int(0.01*len(sortedSents))
topSents = sortedSents[:topPercentageSents]

# print(topSents)
```

```python
summary = []
for sent, strength in sentStrength.items():
    if strength in topSents:
        summary.append(sent)

#for sentence in summary:
    #print(sentence)
    
for sent in doc.sents:
    if sent in summary:
        print(sent)
        

```
